```python
a = float(input("enter side 1: "))
b = float(input("enter side 2: "))
c = float(input("enter side 3: "))
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("equilateral triangle")
    elif a == b or b == c or a == c:
        print("isosceles triangle")
    elif a*a + b*b == c*c or a*a + c*c == b*b or b*b + c*c == a*a:
        print("right-angled triangle")
    else:
        print("scalene triangle")
else:
    print("Not a valid")
```

```python
#13
n = int(input("enter no.:"))
c = 0
for i in range(1, n + 1):
    if n % i == 0:
        c += 1

if c == 2:
    print("prime")
else:
    print("not prime")
```

```python
#14
n = int(input("enter no.:"))
temp = n
s = 0

while n > 0:
    d = n % 10
    f = 1
    for i in range(1, d + 1):
        f *= i
    s += f
    n //= 10

if s == temp:
    print("strong number")
else:
    print("not strong")
```

```python
#15
n = int(input("enter no.:"))
a = 0
b = 1
f = False

while a <= n:
    if a == n:
        f = True
        break
    c = a + b
    a = b
    b = c

if f:
    print("fibonacci")
else:
    print("not fibonacci")
```

```python
#17
n = int(input("enter no."))
def isprime(x):
    if x < 2:
        return False
    c = 0
    for i in range(1, x + 1):
        if x % i == 0:
            c += 1
    return c == 2
rev = 0
temp = n
while temp > 0:
    d = temp % 10
    rev = rev * 10 + d
    temp //= 10

if isprime(n) and isprime(rev):
    print("twisted prime")
else:
    print("not twisted prime")
```

```python
#18
n = int(input("enter no. to check"))
temp = n
rev = 0

while n > 0:
    d = n % 10
    rev = rev * 10 + d
    n //= 10

if rev == temp:
    print("palindrome")
else:
    print("not palindrome")
```
